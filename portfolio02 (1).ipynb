### **SC 5, Question 13:** <a id="q5m"></a>

Your CREATE TABLE statements here from Activity 5.

`````python
CREATE TABLE Company
    (
    COMPANYID INTEGER NOT NULL,
    TITLE CHAR(30),
    PRIMARY KEY(COMPANYID)
    )
    WITHOUT ROWID;
    
CREATE TABLE Product
    (
    PRODUCTID INTEGER NOT NULL,
    TITLE CHAR(30),
    PRIMARY KEY(PRODUCTID)
    )
    WITHOUT ROWID;
    
CREATE TABLE Stock
    (
    COMPANYID INTEGER NOT NULL,
    PRODUCTID INTEGER NOT NULL,
    QUANTITY INTEGER NOT NULL,
    PRIMARY KEY(COMPANYID, PRODUCTID),
    FOREIGN KEY(COMPANYID) REFERENCES Company(COMPANYID),
    FOREIGN KEY(PRODUCTID) REFERENCES Product(PRODUCTID)
    )
    WITHOUT ROWID;
`````

`````python
SELECT DISTINCT M.MVTITLE
    FROM DIRECTOR AS D
    NATURAL JOIN MOVIE AS M
    NATURAL LEFT JOIN BORROW AS B
    NATURAL LEFT JOIN MEMBER AS MB
    WHERE D.DIRNAME = 'Hitchcock, Alfred' AND NOT EXISTS
    (
    SELECT B.MMBNUMB
        WHERE B.MMBNUMB =
        (
        SELECT B.MMBNUMB
            WHERE MB.MMBNAME = 'Tran, Thanh'
        )
    );
`````
